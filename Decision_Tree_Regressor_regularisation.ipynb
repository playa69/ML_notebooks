## Возьем датасет из sklearn и преобразуем его в DataFrame

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
```

```python
data
```

```python
X = data.data
```

```python
X
```

```python
features = data.feature_names
```

```python
features
```

```python
y = data.target
```

```python
df_full = pd.DataFrame(X, columns = features)
```

```python
df_full
```

```python
df_full['target'] = y
df = df_full.iloc[:10]
df
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
   df[features], df['target'], test_size = .2, shuffle = True, random_state=3)
```

```python
from sklearn.tree import DecisionTreeRegressor


tree = DecisionTreeRegressor(random_state=1)
tree.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree


plt.figure(figsize=(20, 15))
plot_tree(tree, feature_names=features, filled=True);
```

```python
from sklearn.metrics import mean_squared_error

pred_train = tree.predict(X_train)
pred_test = tree.predict(X_test)

mse_train = mean_squared_error(y_train, pred_train)
mse_test = mean_squared_error(y_test, pred_test)

print(f'MSE на обучении {mse_train:.2f}')
print(f'MSE на тесте {mse_test:.2f}')
```

## MSE на обучении 0.00 => переобучились

```python
pd.DataFrame({
    'true': y_train,
    'pred': pred_train
})
```

```python
pd.DataFrame({
    'true': y_test,
    'pred': pred_test
})
```

## Уменьшим MAX_DEPTH

```python
tree = DecisionTreeRegressor(random_state=1, max_depth=2)
tree.fit(X_train, y_train)

plt.figure(figsize=(15, 10))
plot_tree(tree, feature_names=features, filled=True);
```

## Увеличим min_samples_leaf

```python
tree = DecisionTreeRegressor(random_state=1, min_samples_leaf=2)
tree.fit(X_train, y_train)

plt.figure(figsize=(15, 10))
plot_tree(tree, feature_names=features, filled=True);
```

## Уменьшим max_leaf_nodes

```python
tree = DecisionTreeRegressor(random_state=1, max_leaf_nodes=5)
tree.fit(X_train, y_train)

plt.figure(figsize=(15, 10))
plot_tree(tree, feature_names=features, filled=True);
```

## Регуляризуем исходую модель

```python
tree = DecisionTreeRegressor(random_state=1,
                             max_depth=2,
                             min_samples_leaf=2,
                             max_leaf_nodes=5)
tree.fit(X_train, y_train)

pred_train = tree.predict(X_train)
pred_test = tree.predict(X_test)
```

```python
mse_train = mean_squared_error(y_train, pred_train)
mse_test = mean_squared_error(y_test, pred_test)

print(f'MSE на обучении {mse_train:.2f}')
print(f'MSE на тесте {mse_test:.2f}')
```
