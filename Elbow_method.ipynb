## делаю три кластера

```python
km3 = sk_KMeans(n_clusters =3, n_init = 'auto').fit(X)
```

```python
km3
```

```python
km3.inertia_
```

```python
inertias = [] #list
for i in range(1, 7):
    dists = sk_KMeans(n_clusters = i, n_init  ="auto").fit(X).inertia_
    inertias.append(dists)
```
